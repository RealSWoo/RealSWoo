```python
import pandas as pd
import numpy as np
import FinanceDataReader as fdr # Finance-DataReader는 Naver와 KRX에서 데이터를 가져옴(외국 주식 등은 야후(yfinance) 등을 이용하여 새로 코드 작성 필요
                                # yfinance처럼 재무제표까지 가져오는 기능은 없음

# 삼성전자 주가 다운로드
df = fdr.DataReader('005930', '2000-01-06', '2025-04-06') # 삼성전자 주식코드 005930

print(df)

# 데이터프레임 개요 확인
df.info()

# 시각화
import matplotlib.pyplot as plt
import seaborn as sns

# seaborn -> lineplot 함수 -> 전체 주가 시각화
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x = df.index, y = 'Close')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()

# 액면분할로 인해서 거래가 중지된 데이터 삭제
# Open 컬럼의 값에 대한 비교 연산
condition = (df.loc[:,'Open']==0)
print(condition)
print('-'*80)

print(f'Open 컬럼의 값이 0인 데이터의 수 : \n{condition.sum()}')

# Open 컬럼의 값에 대한 비교연산 -> 인덱싱 적용
stop = df.loc[df.loc[:,'Open']==0,:]
print(f'거래가 중지된 날의 데이터 : \n{stop}')

print('-'*80)

# Open 컬럼의 값 = 0인 행의 인덱스 추출
stop_index = stop.index
print(f'거래가 중지된 날의 데이터 인덱스 : \n{stop_index}')

# 거래가 중지된 데이터 삭제
df.drop(index = stop_index, inplace = True)
print(df.describe())

# 정답 데이터(target) 생성(close 컬럼 복사)
df.loc[:, 'target'] = df.loc[:, 'Close']
print(df)

# 주가분석(예측) 딥러닝 학습을 위한 입력 데이터(x,y) 생성
X = df.loc[:, 'Open':'Volume']
y = df.loc[:, ['target']]
print(X)
print('-'*80)
print(y)

# 정규화(scaling)
# 딥러닝 학습은 df로는 안되고 numpy 배열로 가능하므로 전처리 마지막에 normalization 해주어야함
# .fit_transform()함수는 numpy로 반환하기 때문에 주의해야함

from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 모델 생성
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

# scaling 실행 : fit_transfrom()
X_scaled = scaler_x.fit_transform(X) # X[i]-min / max-min
y_scaled = scaler_y.fit_transform(y)

print(f'X_scaled의 모양 : {X_scaled.shape}')
print('-'*80)
print(f'X_scaled 확인 : \n{X_scaled}')
print('-'*80)
print(f'y_scaled의 모양 : {y_scaled.shape}')
print('-'*80)
print(f'y_scaled 확인 : \n{y_scaled}')

# 함수 생성
# 1. window_size(time step, 시간 간격) 정의
# (1) 얼마 동안(기간)의 주가 데이터에 기반하여 다음날 종가를 예측할 것인가를 정하는 parameter
# (2) 과거 10일을 기반으로 내일 데이터를 예측한다고 가정했을 때 window_size(시간 간격) = 10 이 된다
# (3) 목표 : 과거 10일을 기준으로 그 다음날의 '종가'를 예측할 수 있는 데이터 생성
# 2. 순차적으로 10일동안 데이터를 하나의 데이터로 묶고 이에 맞는 label(예측 데이터) 생성

# window 사이즈만 바뀌고 나머지는 템플릿으로 사용 가능
def make_dataset(X, y, window_size = 10) :
    X_data_total = []
    y_data_total = []
    for start in np.arange(len(X)-window_size) : # 5996행에서 10개씩 빼서 단위를 만듦
        stop = start + window_size
        # X_data -> 슬라이싱
        X_data = X[start:stop,:] # X_data는 2차원 배열, 그래서 하루 주가 데이터인 행을 10일 간격으로 slicing
        X_data_total.append(X_data)
        # y_data -> 인덱싱
        y_data = y[stop] # y데이터는 y[10]으로 인덱스 10번 데이터 할당
        y_data_total.append(y_data)
    return np.array(X_data_total), np.array(y_data_total) # 학습은 np 배열 형태만 가능, 리스트 형태는 학습 불가

# 시계열 분석에 적합한 데이터 생성
# 데이터 생성 함수 실행
X_data, y_data = make_dataset(X=X_scaled, y=y_scaled)

print(f'X_data의 모양 : {X_data.shape}')
print('-'*80)
print(f'첫 번째 X_data 확인 : \n{X_data[0,:,:]}')
print('-'*80)
print(f'y_data의 모양 : {y_data.shape}')
print(f'첫 번째 y_data 확인 : \n{y_data[0]}')

print(df.describe())


# 학습용 / 검증용 / 평가용 데이터 생성 2 - 오차를 줄이기 위해 랜덤하게 데이터를 분할
# TEST_SIZE 설정
# (1) TEST_SIZE = 200으로 설정
# (2) TEST_SIZE = 200은 가장 최근 200일의 데이터로 모델이 주가를 예측
# (3) 학습은 과거부터 200일 이전의 데이터를 학습

TEST_SIZE = 200
# 학습용 데이터 생성
X_train = X_data[0:-TEST_SIZE, :, :]
y_train = y_data[0:-TEST_SIZE, :]
# 평가용 데이터 생성
X_test = X_data[-TEST_SIZE:, :, :]
y_test = y_data[-TEST_SIZE:, :]

from sklearn.model_selection import train_test_split

# 학습용 데이터 분할
X_train, X_val, y_train, y_val = train_test_split (X_train,
                                                   y_train,
                                                   test_size=0.2,
                                                   random_state=0) # 랜덤하게 구성됨

# 학습용 / 검증용 / 평가용 데이터의 모양 확인 2
print(f'학습용 X 데이터의 모양 : {X_train.shape}')
print('-'*80)
print(f'학습용 y 데이터의 모양 : {y_train.shape}')
print('-'*80)
print(f'검증용 X 데이터의 모양 : {X_val.shape}')
print('-'*80)
print(f'검증용 X 데이터의 모양 : {y_val.shape}')
print('-'*80)
print(f'평가용 X 데이터의 모양 : {X_test.shape}')
print('-'*80)
print(f'평가용 X 데이터의 모양 : {y_test.shape}')

# 랜덤 시드 설정
import tensorflow as tf
import random
import numpy as np

random.seed(0)
np.random.seed(0)
tf.random.set_seed(0)
initializer1 = tf.keras.initializers.GlorotUniform(seed=0)
initializer2 = tf.keras.initializers.Orthogonal(seed=0)

# 모델 구조 정의
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(units=128,
                               kernel_initializer=initializer1,
                               recurrent_initializer=initializer2,
                               input_shape=(10,5)))

model.add(tf.keras.layers.Dense(units=1,
                                kernel_initializer=initializer1))
model.summary()

# 모델 컴파일
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='mean_squared_error', optimizer=optimizer)

# 모델 학습 2
earlystop = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    patience = 5,
    verbose = 1
)

# 최적의 학습 결과를 저장하기 위한 ModelCheckpoint 설정
file_path = '/home/user-a/Desktop/.h5'
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath = file_path,
    monitor = 'val_loss',
    verbose = 1,
    save_best_only = True
)

# 학습 진행
model.fit(x=X_train,
          y=y_train,
          batch_size=256,
          validation_data=(X_val,y_val),
          epochs=200,
          callbacks=[earlystop, checkpoint]) # train_test_plit_ 했기 때문에, validation_split 빠짐 -> validation_data 추가함
prediction = model.predict(X_test)
print(f'최근 200일 동안 삼성전자의 주가에 대한 예측 : \n{prediction}')

# 성능 개선 후 평가용 데이터에 대한 예측의 결과 --> 데이터프레임 생성 2
index = y.iloc[-200:].index
df_test = pd.DataFrame(data=y_test, index=index, columns=['y_test'])
df_test.loc[:, 'prediction'] = prediction
print(df_test)

pred_inverse = scaler_y.inverse_transform(prediction)
print(f'원래 가격으로 변환시킨 삼성전자 주가 예측 값 : \n{pred_inverse}')

# 성능 평가 후 최근 200일의 실제 주가와 스케일링 이전의 상태로 변환된 예측값 -> df_test에 추가 2
df_test.loc[:, 'test_origin'] = y.iloc[-200:, :].values
print(df_test)
print('-'*80)

df_test.loc[:, 'pred_origin'] = pred_inverse # 주가 원래가격(스케일링 전)
print(df_test)
print('-'*80)

# error 계산 -> df_test에 새로운 컬럼으로 추가
# feature engineering (기존 컬럼을 가지고 새로운 컬럼을 만들어냄 = 파생시킨다. 정답 - 예측값)
error = df_test.loc[:, 'test_origin'] - df_test.loc[:, 'pred_origin']
df_test.loc[:, 'error'] = error
display(df_test)

# 성능 개선 후 결과 시각화 2
plt.figure(figsize=(12,6))
df_test.loc[:, 'test_origin':'pred_origin'].plot(kind='line')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```
