```python
import pandas as pd
import numpy as np
import FinanceDataReader as fdr # Finance-DataReader Naver KRX  (   (yfinance)      
                                # yfinance    

#   
df = fdr.DataReader('005930', '2000-01-06', '2025-04-06') #   005930

print(df)

#   
df.info()

# 
import matplotlib.pyplot as plt
import seaborn as sns

# seaborn -> lineplot  ->   
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x = df.index, y = 'Close')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()

#      
# Open     
condition = (df.loc[:,'Open']==0)
print(condition)
print('-'*80)

print(f'Open   0   : \n{condition.sum()}')

# Open     ->  
stop = df.loc[df.loc[:,'Open']==0,:]
print(f'    : \n{stop}')

print('-'*80)

# Open   = 0   
stop_index = stop.index
print(f'     : \n{stop_index}')

#    
df.drop(index = stop_index, inplace = True)
print(df.describe())

#  (target) (close  )
df.loc[:, 'target'] = df.loc[:, 'Close']
print(df)

# ()     (x,y) 
X = df.loc[:, 'Open':'Volume']
y = df.loc[:, ['target']]
print(X)
print('-'*80)
print(y)

# (scaling)
#   df  numpy     normalization 
# .fit_transform() numpy   

from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler  
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

# scaling  : fit_transfrom()
X_scaled = scaler_x.fit_transform(X) # X[i]-min / max-min
y_scaled = scaler_y.fit_transform(y)

print(f'X_scaled  : {X_scaled.shape}')
print('-'*80)
print(f'X_scaled  : \n{X_scaled}')
print('-'*80)
print(f'y_scaled  : {y_scaled.shape}')
print('-'*80)
print(f'y_scaled  : \n{y_scaled}')

#  
# 1. window_size(time step,  ) 
# (1)  ()         parameter
# (2)  10       window_size( ) = 10  
# (3)  :  10    ''     
# 2.  10       label( ) 

# window      
def make_dataset(X, y, window_size = 10) :
    X_data_total = []
    y_data_total = []
    for start in np.arange(len(X)-window_size) : # 5996 10   
        stop = start + window_size
        # X_data -> 
        X_data = X[start:stop,:] # X_data 2 ,      10  slicing
        X_data_total.append(X_data)
        # y_data -> 
        y_data = y[stop] # y y[10]  10  
        y_data_total.append(y_data)
    return np.array(X_data_total), np.array(y_data_total) #  np   ,    

#     
#    
X_data, y_data = make_dataset(X=X_scaled, y=y_scaled)

print(f'X_data  : {X_data.shape}')
print('-'*80)
print(f'  X_data  : \n{X_data[0,:,:]}')
print('-'*80)
print(f'y_data  : {y_data.shape}')
print(f'  y_data  : \n{y_data[0]}')

print(df.describe())


#  /  /    2 -      
# TEST_SIZE 
# (1) TEST_SIZE = 200 
# (2) TEST_SIZE = 200   200    
# (3)   200   

TEST_SIZE = 200
#   
X_train = X_data[0:-TEST_SIZE, :, :]
y_train = y_data[0:-TEST_SIZE, :]
#   
X_test = X_data[-TEST_SIZE:, :, :]
y_test = y_data[-TEST_SIZE:, :]

from sklearn.model_selection import train_test_split

#   
X_train, X_val, y_train, y_val = train_test_split (X_train,
                                                   y_train,
                                                   test_size=0.2,
                                                   random_state=0) #  

#  /  /     2
print(f' X   : {X_train.shape}')
print('-'*80)
print(f' y   : {y_train.shape}')
print('-'*80)
print(f' X   : {X_val.shape}')
print('-'*80)
print(f' X   : {y_val.shape}')
print('-'*80)
print(f' X   : {X_test.shape}')
print('-'*80)
print(f' X   : {y_test.shape}')

#   
import tensorflow as tf
import random
import numpy as np

random.seed(0)
np.random.seed(0)
tf.random.set_seed(0)
initializer1 = tf.keras.initializers.GlorotUniform(seed=0)
initializer2 = tf.keras.initializers.Orthogonal(seed=0)

#   
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(units=128,
                               kernel_initializer=initializer1,
                               recurrent_initializer=initializer2,
                               input_shape=(10,5)))

model.add(tf.keras.layers.Dense(units=1,
                                kernel_initializer=initializer1))
model.summary()

#  
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='mean_squared_error', optimizer=optimizer)

#   2
earlystop = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    patience = 5,
    verbose = 1
)

#      ModelCheckpoint 
file_path = '/home/user-a/Desktop/.h5'
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath = file_path,
    monitor = 'val_loss',
    verbose = 1,
    save_best_only = True
)

#  
model.fit(x=X_train,
          y=y_train,
          batch_size=256,
          validation_data=(X_val,y_val),
          epochs=200,
          callbacks=[earlystop, checkpoint]) # train_test_plit_  , validation_split  -> validation_data 
prediction = model.predict(X_test)
print(f' 200      : \n{prediction}')

#         -->   2
index = y.iloc[-200:].index
df_test = pd.DataFrame(data=y_test, index=index, columns=['y_test'])
df_test.loc[:, 'prediction'] = prediction
print(df_test)

pred_inverse = scaler_y.inverse_transform(prediction)
print(f'       : \n{pred_inverse}')

#     200        -> df_test  2
df_test.loc[:, 'test_origin'] = y.iloc[-200:, :].values
print(df_test)
print('-'*80)

df_test.loc[:, 'pred_origin'] = pred_inverse #  ( )
print(df_test)
print('-'*80)

# error  -> df_test   
# feature engineering (      = .  - )
error = df_test.loc[:, 'test_origin'] - df_test.loc[:, 'pred_origin']
df_test.loc[:, 'error'] = error
display(df_test)

#      2
plt.figure(figsize=(12,6))
df_test.loc[:, 'test_origin':'pred_origin'].plot(kind='line')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```
