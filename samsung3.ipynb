```python
# 1.    
import pandas as pd
import numpy as np
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
import random

#   ( )
random.seed(0)
np.random.seed(0)
tf.random.set_seed(0)

# 2.  +   
df_stock = fdr.DataReader('005930', '2000-01-06', '2025-04-06')
df_fx = fdr.DataReader('USD/KRW', '2000-01-06', '2025-04-06')[['Close']]
df_fx.rename(columns={'Close': 'USD/KRW'}, inplace=True)

# 3.  
df = df_stock.merge(df_fx, left_index=True, right_index=True, how='inner')
df = df[df['Open'] != 0]  #   
df['target'] = df['Close']

# 4.  
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='USD/KRW', y='Close', alpha=0.3)
plt.title("USD/KRW  vs  ")
plt.xlabel("USD/KRW ")
plt.ylabel(" ")
plt.grid(True)
plt.show()

corr = df['USD/KRW'].corr(df['Close'])
print(f'/    : {corr:.4f}')

# 5. /   
X = df[['Open', 'High', 'Low', 'Close', 'Volume', 'USD/KRW']]
y = df[['target']]

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# 6.    
def make_dataset(X, y, window_size=10):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        Xs.append(X[i:i+window_size])
        ys.append(y[i+window_size])
    return np.array(Xs), np.array(ys)

X_data, y_data = make_dataset(X_scaled, y_scaled)

# 7. // 
TEST_SIZE = 200
X_train_all, X_test = X_data[:-TEST_SIZE], X_data[-TEST_SIZE:]
y_train_all, y_test = y_data[:-TEST_SIZE], y_data[-TEST_SIZE:]

X_train, X_val, y_train, y_val = train_test_split(X_train_all, y_train_all,
                                                  test_size=0.2, random_state=0)

# 8.  
initializer1 = tf.keras.initializers.GlorotUniform(seed=0)
initializer2 = tf.keras.initializers.Orthogonal(seed=0)

model = tf.keras.Sequential([
    tf.keras.layers.LSTM(128,
                         kernel_initializer=initializer1,
                         recurrent_initializer=initializer2,
                         input_shape=(10, 6)),
    tf.keras.layers.Dense(1, kernel_initializer=initializer1)
])

model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001))

#  
earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, verbose=1)
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath='best_model.h5',
                                                monitor='val_loss',
                                                save_best_only=True,
                                                verbose=1)

# 9. 
model.fit(X_train, y_train,
          epochs=200,
          batch_size=256,
          validation_data=(X_val, y_val),
          callbacks=[earlystop, checkpoint])

# 10.   
prediction = model.predict(X_test)
pred_origin = scaler_y.inverse_transform(prediction)
y_test_origin = scaler_y.inverse_transform(y_test)

# 11.    
df_test = pd.DataFrame({
    'test_origin': y_test_origin.flatten(),
    'pred_origin': pred_origin.flatten()
}, index=df.iloc[-TEST_SIZE:].index)

df_test['error'] = df_test['test_origin'] - df_test['pred_origin']

#  (  )
plt.figure(figsize=(12, 6))
plt.plot(df_test.index, df_test['test_origin'], label=' ', linewidth=2)
plt.plot(df_test.index, df_test['pred_origin'], label=' ', linewidth=2, linestyle='--')
plt.title("   vs  ( )")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 12.   
print(df_test.describe())
```
